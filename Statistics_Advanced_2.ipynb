1. What is hypothesis testing in statistics ?

- Hypothesis testing is a statistical procedure where a null hypothesis (an assumption about a population) is tested against an alternative hypothesis using sample data to determine if there's enough evidence to reject the null hypothesis. The goal is to infer properties of a whole population based on a sample, distinguishing between genuine effects and those that could have occurred by random chance. This structured, data-driven process helps researchers validate theories and make informed decisions about population-level claims.  

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

- The null hypothesis (H₀) states that there is no significant effect, relationship, or difference between variables, while the alternative hypothesis (H₁ or Hₐ) proposes that a statistically significant effect, relationship, or difference does exist. The main difference is that the null hypothesis is a statement of the default or baseline assumption of no effect, which the researcher tries to disprove, whereas the alternative hypothesis is the researcher's prediction or claim that they are trying to prove.

Null Hypothesis (H₀)

What it is: A default, conservative statement that there is no real effect or relationship between the variables being studied.
Purpose: It is the hypothesis that the researcher aims to reject or disprove with their data.

Symbol: Always denoted as H₀.

Mathematical representation: Often uses an equals sign (=), or signs like greater than or equal to (≥) or less than or equal to (≤).

Example: "There is no difference in the average test scores between students who use study method A and students who use study method B".

Alternative Hypothesis (H₁ or Hₐ)

What it is: A statement that proposes that there is a statistically significant relationship or effect between variables.

Purpose: This is the hypothesis the researcher aims to prove or establish.
Symbol: Denoted as H₁ or Hₐ.

Mathematical representation: Uses inequality symbols (<, >, or ≠).
Example: "Students who use study method A will have significantly higher test scores than students who use study method B".

Key Differences Summarized

Direction: H₀ asserts "no effect," while H₁ asserts "there is an effect".

Researcher's Goal: Researchers work to disprove H₀ and support H₁.

Starting Point: H₀ is the default assumption that is believed to be true until proven false.

Conclusion: If the null hypothesis is rejected, it means there is evidence to support the alternative hypothesis.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

- The significance level (α) in hypothesis testing sets the threshold for rejecting the null hypothesis, representing the maximum allowable probability of a Type I error (incorrectly rejecting a true null hypothesis). To decide a test's outcome, the calculated p-value is compared to the pre-determined significance level: if p < α, the null hypothesis is rejected; if p ≥ α, it is not rejected, indicating the results are not statistically significant at that level.

What is the Significance Level (α)?

Pre-determined Threshold: The significance level, denoted by the Greek letter alpha (α), is a threshold set by the researcher before the test is conducted.

Maximum Risk of Type I Error: It represents the maximum probability of making a Type I error – falsely concluding that the null hypothesis is false when it is actually true.

Commonly Used Values: Common values for α are 0.05 (5%), 0.01 (1%), and 0.10 (10%), though other values can be used depending on the study's context and costs.

Role in Deciding the Outcome

The significance level acts as a benchmark to interpret the results of a hypothesis test.

Calculate the Test Statistic and p-value: After collecting data, a test statistic is calculated from the sample data. This statistic is then used to find the p-value.

Compare p-value to α: The calculated p-value is compared to the pre-set significance level (α).

If p < α: The null hypothesis is rejected. This means the observed data is unlikely to have occurred if the null hypothesis were true, leading to the conclusion that the results are statistically significant.

If p ≥ α: The null hypothesis is not rejected. This indicates that the data is consistent with the null hypothesis, and there is not enough evidence to conclude that the null hypothesis is false.

Why it's Significant

Controls Error Rate: It helps control the rate of false positives in research, ensuring that only a small, pre-determined proportion of true null hypotheses are rejected.

Provides a Decision Rule: It provides a clear decision rule for interpreting the evidence gathered from the data.

Avoids Data Dredging: Establishing the significance level before data analysis helps prevent p-hacking, a practice of manipulating data or methods to achieve a desired p-value.

4. What are Type I and Type II errors? Give examples of each.

- A Type I error is a "false positive," where a true null hypothesis is incorrectly rejected, while a Type II error is a "false negative," where a false null hypothesis is incorrectly failed to be rejected. Examples include a Type I error of a fire alarm ringing when there is no fire, and a Type II error of a medical test failing to detect a disease that is actually present.  

Type I Error (False Positive)

Definition: A Type I error occurs when you reject the null hypothesis (H₀) even though it is actually true.

Symbol: The probability of a Type I error is denoted by the Greek letter alpha (α).

Analogy: Think of it as a "false alarm" or "crying wolf".

Examples:

Medical Testing: A patient tests positive for a disease, but they do not actually have the disease.

Criminal Justice: A court finds an innocent person guilty of a crime.
Manufacturing: A quality control inspector incorrectly concludes that a product is defective when it is not.

Type II Error (False Negative)

Definition: A Type II error occurs when you fail to reject the null hypothesis (H₀) even though it is false.

Symbol: The probability of a Type II error is denoted by the Greek letter beta (β).

Analogy: Think of it as "missing the target" or a "failed detection".

Examples:

Medical Testing: A medical test fails to detect a disease, but the patient actually has it.

Criminal Justice: A court finds a guilty person not guilty of the crime.

Product Testing: A product is defective, but the quality control inspector fails to detect the defect.

Key Difference Summary

Type I Error (α): A true null hypothesis is rejected (False Positive).
Type II Error (β): A false null hypothesis is not rejected (False Negative).

5. What is the difference between a Z-test and a T-test? Explain when to use each.

- A Z-test is used when the population standard deviation is known and the sample size is large (typically n ≥ 30), while a T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30). The main difference is the knowledge of the population's standard deviation and the corresponding test statistic and distribution used (standard normal distribution for Z-tests, and the t-distribution for T-tests).

When to Use a Z-Test

You should use a Z-test when the following conditions are met:

Population Standard Deviation is Known: You have information about the standard deviation of the entire population from which your sample was drawn.

Large Sample Size: Your sample size is sufficiently large, usually considered to be 30 or more.

Normal Distribution: The data from which the sample was drawn is assumed to be normally distributed, or your sample is large enough that the Central Limit Theorem applies.

When to Use a T-Test

You should use a T-test when the following conditions are met:

Population Standard Deviation is Unknown: The standard deviation of the population is not known, and you must estimate it using the sample's standard deviation.

Small Sample Size: Your sample size is small, generally less than 30.
Normal Distribution: The data is approximately normally distributed.

```python
#8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simulate_and_plot_confidence_interval(sample_size=1000, true_mean=50, true_std=10, confidence_level=0.95):


    np.random.seed(0)
    data = np.random.normal(true_mean, true_std, sample_size)


    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = stats.sem(data)


    degrees_freedom = sample_size - 1
    alpha = 1 - confidence_level
    t_critical = stats.t.ppf(1 - alpha / 2, degrees_freedom)
    margin_of_error = t_critical * standard_error

    confidence_interval_low = sample_mean - margin_of_error
    confidence_interval_high = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"95% Confidence Interval: ({confidence_interval_low:.2f}, {confidence_interval_high:.2f})")


    plt.figure(figsize=(10, 6))


    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Data Histogram')


    plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')
    plt.axvline(confidence_interval_low, color='green', linestyle='dashed', linewidth=2, label=f'95% CI Lower Bound ({confidence_interval_low:.2f})')
    plt.axvline(confidence_interval_high, color='green', linestyle='dashed', linewidth=2, label=f'95% CI Upper Bound ({confidence_interval_high:.2f})')


    plt.title(f'Simulated Normal Data and {int(confidence_level*100)}% Confidence Interval')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(axis='y', alpha=0.75)
    plt.show()


simulate_and_plot_confidence_interval()
```

```python
#9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_visualize_zscores(data):

    data = np.array(data)


    mean = np.mean(data)
    std_dev = np.std(data)


    z_scores = (data - mean) / std_dev

    print(f"Original Data: {data}")
    print(f"Mean: {mean:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Calculated Z-scores: {z_scores}")


    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

sample_data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 35]
calculate_and_visualize_zscores(sample_data)
```
